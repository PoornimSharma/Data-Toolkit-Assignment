### Q1- What is NumPy, and why is it widely used in Python

NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides a powerful N-dimensional array object, along with tools for integrating with C/C++ and Fortran code. NumPy is widely used because it enables efficient numerical operations on large datasets. Its core advantage is that its arrays are much faster and more memory-efficient than standard Python lists, which makes it an essential tool for data science, machine learning, and scientific research.

***

### Q2- How does broadcasting work in NumPy

Broadcasting is a powerful feature in NumPy that allows arrays with different shapes to be used in arithmetic operations. It works by "stretching" the smaller array across the larger one so that they have compatible shapes for element-wise operations. This eliminates the need to manually create multiple copies of the smaller array. For broadcasting to work, the trailing dimensions (or size 1 dimensions) of the arrays must be compatible. For example, a scalar can be broadcast to an array, and a 1D array can be broadcast to a 2D array.

### Q3- What is a Pandas DataFrame

A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). Think of it as a spreadsheet or SQL table in Python. It's the primary data structure in Pandas and is widely used for data manipulation and analysis. DataFrames are designed to make data cleaning, transformation, and analysis tasks both efficient and easy by providing a rich set of built-in functions for handling various data types, missing values, and complex operations.

### Q4- Explain the use of the groupby() method in Pandas

The `groupby()` method in Pandas is used to split a DataFrame into groups based on some criterion. It follows a "split-apply-combine" pattern:
1.  **Split**: Data is grouped into subsets based on one or more keys.
2.  **Apply**: A function (e.g., `mean()`, `sum()`, `count()`) is applied independently to each group.
3.  **Combine**: The results of the applied function are combined into a single new DataFrame or Series.
This method is essential for performing aggregate analysis on subsets of data, such as finding the average sales for each region.

### Q7- What is a heatmap, and when should it be used

A heatmap is a graphical representation of data where values are depicted by color. It's excellent for visualizing the magnitude of a phenomenon, such as a correlation matrix. The color scale can be continuous or discrete, with different shades representing different value ranges. ### Q8- What does the term “vectorized operation” mean in NumPy

A vectorized operation in NumPy refers to applying an operation to an entire array at once, rather than iterating through its elements using a Python loop. For example, `array_a + array_b` performs an element-wise addition on two arrays without explicit looping. This is a core feature of NumPy. Vectorized operations are implemented in optimized C code, making them significantly faster and more memory-efficient than their pure Python loop equivalents. This approach is what gives NumPy its high performance for numerical computations.

### Q15- How does NumPy handle multidimensional arrays

NumPy handles multidimensional arrays by treating them as a single, contiguous block of memory with a metadata structure that describes their shape and stride (the number of bytes to jump to get to the next element in a dimension). For example, a 2D array is stored as a 1D array, and NumPy's indexing mechanism calculates the correct memory location for each element. This efficient memory layout allows for very fast access and vectorized operations across multiple dimensions, making it a cornerstone for scientific computing.

### Q17- Explain the difference between apply() and map() in Pandas

`map()` is a Series method that is used to substitute each value in a Series with another value, often by using a dictionary or a function. It's designed for element-wise mapping and works only on a single Series. `apply()` is a more versatile method that can be used on both a Series and a DataFrame. When used on a DataFrame, it applies a function along a specific axis (row or column). `apply()` is generally slower than `map()` because it's more flexible and often requires Python loops under the hood, but it's essential for more complex row-wise or column-wise operations.

### Q18- What are some advanced features of NumPy

NumPy's advanced features include:
1.  **Broadcasting**: Allows operations on arrays with different shapes.
2.  **Linear Algebra Functions**: Provides functions for operations like matrix decomposition, determinants, and solving linear equations.

### Q20- What is the role of a pivot table in Pandas

***

### Q21- Why is NumPy’s array slicing faster than Python’s list slicing

NumPy's array slicing is faster than Python's list slicing because it works by creating a **view** of the original array, which is a new array object that points to the same data in memory. This is a very fast and memory-efficient process. In contrast, Python's list slicing creates a **new copy** of the list in a different memory location. This copying process takes more time and memory, especially for large lists, making NumPy's slicing significantly more performant for large-scale data manipulation.

```python
# Q4- How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {'A': np.random.rand(10),
        'B': np.random.rand(10) + np.random.rand(10) * 0.5,
        'C': np.random.rand(10) - np.random.rand(10) * 0.5}
df = pd.DataFrame(data)

# Compute the correlation matrix
correlation_matrix = df.corr()

print("Correlation Matrix:\n", correlation_matrix)

import matplotlib.pyplot as plt
import seaborn as sns

# Create the heatmap
plt.figure(figsize=(8, 6)) # Set the size of the plot
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Q5- Generate a bar plot using Plotly?

import plotly.graph_objects as go

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [250, 450, 300, 520]

# Create a bar trace
bar_chart = go.Bar(x=products, y=sales)

# Create a figure and add the bar chart
fig = go.Figure(data=[bar_chart])

# Update layout for titles and labels
fig.update_layout(
    title='Quarterly Sales by Product',
    xaxis_title='Product',
    yaxis_title='Sales (in units)'
)

# Display the plot
fig.show()
```

```python
# Q8- Create a line plot with multiple lines using Matplotlib?


import matplotlib.pyplot as plt
import numpy as np

# Sample data for three different lines
x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Plot each line
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) * cos(x)')

# Add a title and labels
plt.title('Multiple Lines on a Single Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend to distinguish the lines
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Q11- Perform matrix multiplication using NumPy?

import numpy as np

# Create two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication using the @ operator
C = A @ B

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nResult of A @ B:\n", C)
```

```python
# Q13- Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go
import numpy as np

# Create sample data for the x, y, and z axes
n = 100 # Number of data points
t = np.linspace(-10, 10, n)
x_data = np.sin(t)
y_data = np.cos(t)
z_data = t

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x_data,
    y=y_data,
    z=z_data,
    mode='markers',
    marker=dict(
        size=5,
        color=z_data,  # Color points based on the z-value
        colorscale='Viridis', # Choose a colorscale
        opacity=0.8
    )
)])

# Update the layout for better visualization
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Display the plot
fig.show()
```
